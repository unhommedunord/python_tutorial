```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="example")

l = {'REGION': ['Toronto','Vaughan','Vancouver','Montreal']}

l['COORDS'] = []
for k in l['REGION']:
    loc = geolocator.geocode(k).raw
    l['COORDS'].append((loc['lat'], loc['lon']))

df = pd.DataFrame(l)

map_audience = folium.Map(location=[43.0896, -79.0849], zoom_start=5)

for i,r in df.iterrows():
    folium.Marker(location=r['COORDS'],
                  popup = r['REGION'],
                  tooltip = r['REGION']).add_to(map_audience)
map_audience
```

```python
map_audience.save('map_audience.html')
```

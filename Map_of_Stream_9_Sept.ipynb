```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="example")

l = {'REGION': ['Moscow','New-York','Magnitogorsk','Toronto','Gorlinka','Ekaterinburg','Novorossiisk','Brooklyn','Ashgabat','Madrid','Stockholm','Московская область','Lithuania','Voronezh','Frankfurt','Sankt-Peterburg','Rostov-Na-Donu','Amsterdam','Белгородская область','Latvia','Milan','Dubna','Kemerovo','San Diego','Minsk','Винница','Тверь','Варшава','Уфа','Тбилиси','Пенза','Швеция','Haifa','Касимов','Калининград','Мюнхен','Самара','Краснодар','Кёльн','Прага','Пенза','Иркутск','San Francisco','Санкт-Петербург','Tarzana','Череповец','Одесса','Пермь','Актау','Bensonhurst','Рига','Турция','Воронеж','Сортавала','Киров','Florida','Reno','Барнаул','Norway','Финляндия','Турне','Бельгия','Люксембург','Саратов','Балыкесир','Грузия','Нетания','Чикаго','Донецк','Львов','Bergen','Обливская','Ижевск','Новосибирск','Skokie','Дилижан','Первоуральск','Валенсия','Вильнюс','Аликанте','Астрахань','Киев','Турку','Лахта','Чита','Саратовская область','Ивановская область','Харьков','Израиль','Северодвинск']}

l['COORDS'] = []
for k in l['REGION']:
    loc = geolocator.geocode(k).raw
    l['COORDS'].append((loc['lat'], loc['lon']))

df = pd.DataFrame(l)

map_audience = folium.Map(location=[65,26], zoom_start=4)

for i,r in df.iterrows():
    folium.Marker(location=r['COORDS'],
                  popup = r['REGION'],
                  tooltip = r['REGION']).add_to(map_audience)
map_audience
```

```python
map_audience.save('map_9Sept2022.html')
```
